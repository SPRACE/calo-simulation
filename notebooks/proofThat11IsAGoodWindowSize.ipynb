```python
import matplotlib.pyplot as plt
import seaborn; seaborn.set_style('whitegrid')
```

```python
X = []
Y = []
eY = []
with open("../data/ele100GeV_energyVsSizeOfWindow.txt") as f:
    lines = f.readlines()
    #print(lines)
    for line in lines:
        X.append(float(line.split()[0]))
        Y.append(float(line.split()[4]))
        eY.append(float(line.split()[6]))
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.errorbar(X, Y, yerr=eY, fmt='o', ecolor='orangered',
            color='steelblue', capsize=2)
plt.ylim([84, 88])
plt.xlabel("Window size [squares]",fontsize=18)
plt.ylabel("RECO energy [GeV]",fontsize=18)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.errorbar(X, Y, yerr=eY, fmt='o', ecolor='orangered',
            color='steelblue', capsize=2)
plt.ylim([0, 110])
plt.xlabel("Window size [squares]",fontsize=18)
plt.ylabel("RECO energy [GeV]",fontsize=18)
```
